# Morphological Analysis 👇

```python
word_lemitizer=WordNetLemmatizer()
```

```python
Regular_expression_definition_for_html_tags=re.compile('<.*?>')
Regular_expression_definition_for_digits=re.compile('\d*\s')
```

```python
Regular_expression_definition_for_html_tags
```

```python
Regular_expression_definition_for_html_tags.sub(r" ","<     h1> This is a h1 tag </h1>").strip()
```

```python
english_stop_words=stopwords.words('english')
```

```python
english_stop_words
```

```python
english_stop_words.index("the")
```

## This is used to compare the words before and after the lemmitization

```python
def lemmetize_comparer(text):
    text=text.lower().split()
    text=[word for word in text if word not in english_stop_words]
    text_=[word_lemitizer.lemmatize(word) for word in text]
    for i in range(len(text_)):
        print(text[i]," ->",text_[i])
```

```python
lemmetize_comparer(data[review_column][100])
```

## This Function will preprocess thegiven sentence 
process on each word 

converts in to lower case

romoves stop words

uses lemmitizer on each word

```python
syns = wordnet.synsets("video")
```

```python
syns
```

```python
Movie_related_Keyword=[
    'picture',
    'movie',
    'video',
    'film',
    'biography',
    'cinema',
    'hero',
    'heroine',
    'actor',
    'actress',
    'character'
]
def preprocessing_of_sentence(text):
    text=Regular_expression_definition_for_html_tags.sub(r" ",text)
    text=Regular_expression_definition_for_digits.sub(r" ",text)
    punctuations = [".",",","!","?","'",'"',":",";","*","-","/","+","%","$","#","@","(",")","[","]","{","}"]
    for i in punctuations:
        text = text.replace(i," ")
    text=text.lower().split()
    text=[word for word in text if word not in english_stop_words and len(word)>1 and word not in Movie_related_Keyword]
    text=[word_lemitizer.lemmatize(word) for word in text]
    return text
```

```python
preprocessing_of_sentence(data[review_column][100])
```

```python
len(preprocessing_of_sentence(data[review_column][100]))
```

```python
Preprocessed_textual_data=[" ".join(preprocessing_of_sentence(text)) for text in data[review_column]]
```
